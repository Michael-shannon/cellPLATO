```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from fastdtw import fastdtw
import matplotlib.pyplot as plt

# Generate example DataFrame with columns 'uniq_id', 'frame', and 'label'
# Replace this with your actual DataFrame
# data = {
#     'uniq_id': np.repeat(np.arange(1, 6), 10),
#     'frame': np.tile(np.arange(1, 11), 5),
#     'label': [list(np.random.randint(1, 4, size=50)) for _ in range(50)]  # Keep as-is
# }
# df = pd.DataFrame(data)

df = tptlabel_dr_df
df['label'] = df['label'].apply(lambda x: np.array(x).reshape(-1, 1))
# show the new label column of the df
print(df['label'])
# Calculate pairwise DTW distances
uniq_ids = df['uniq_id'].unique()
distance_matrix = np.zeros((len(uniq_ids), len(uniq_ids)))

for i, uniq_id1 in enumerate(uniq_ids):
    for j, uniq_id2 in enumerate(uniq_ids):
        if i <= j:
            sequences1 = df[df['uniq_id'] == uniq_id1]['label']
            sequences2 = df[df['uniq_id'] == uniq_id2]['label']
            if uniq_id1 == uniq_id2:
                distance_matrix[i, j] = 0  # DTW distance with itself is 0
            else:
                min_distance = np.inf
                for seq1 in sequences1:
                    for seq2 in sequences2:
                        distance, _ = fastdtw(seq1, seq2)
                        min_distance = min(min_distance, distance)
                distance_matrix[i, j] = min_distance
                distance_matrix[j, i] = min_distance

# Perform hierarchical clustering
linkage_matrix = linkage(distance_matrix, method='ward')

# Set a distance threshold to define clusters
threshold = 10  # Adjust as needed
cluster_labels = fcluster(linkage_matrix, t=threshold, criterion='distance')

# Create a dictionary to map uniq_id to trajectory_id
trajectory_mapping = dict(zip(uniq_ids, cluster_labels))

# Add 'trajectory_id' to the DataFrame based on uniq_id
df['trajectory_id'] = df['uniq_id'].map(trajectory_mapping)

# Visualize the dendrogram (optional)
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix, labels=uniq_ids)
plt.xlabel('uniq_id')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()

# Print DataFrame with trajectory IDs
print("DataFrame with Trajectory IDs:")
print(df[['uniq_id', 'frame', 'label', 'trajectory_id']])
```

```python
# HDBSCAN VERSION - do HDBSCAN UNTIL IT FINDS AT LEAST 2 CLUSTERS

import numpy as np
import pandas as pd
from fastdtw import fastdtw
import hdbscan
import matplotlib.pyplot as plt
import tqdm as tqdm

# Generate example DataFrame with columns 'uniq_id', 'frame', and 'label'
# Replace this with your actual DataFrame
# data = {
#     'uniq_id': np.repeat(np.arange(1, 6), 10),
#     'frame': np.tile(np.arange(1, 11), 5),
#     'label': [list(np.random.randint(1, 4, size=50)) for _ in range(50)]  # Keep as-is
# }
# df = pd.DataFrame(data)

df = tptlabel_dr_df
# df['label'] = df['label'].apply(lambda x: list(x))
# df['label'] = df['label'].apply(lambda x: list(x) if not isinstance(x, list) else x)
df['label'] = df['label'].apply(lambda x: np.array(x).reshape(-1, 1))
# show the new label column of the df
print(df['label'])

# Initialize variables for clustering
cluster_labels = np.array([-1] * len(df))
n_clusters = 0
max_iterations = 5
iteration = 0

# Keep clustering until at least two clusters are detected
while iteration < max_iterations and n_clusters < 2:
    # Calculate pairwise DTW distances
    uniq_ids = df['uniq_id'].unique()
    distance_matrix = np.zeros((len(uniq_ids), len(uniq_ids)))

    for i, uniq_id1 in tqdm.tqdm(enumerate(uniq_ids)):
        for j, uniq_id2 in enumerate(uniq_ids):
            if i <= j:
                sequences1 = df[df['uniq_id'] == uniq_id1]['label']
                sequences2 = df[df['uniq_id'] == uniq_id2]['label']
                if uniq_id1 == uniq_id2:
                    distance_matrix[i, j] = 0  # DTW distance with itself is 0
                else:
                    min_distance = np.inf
                    for seq1 in sequences1:
                        for seq2 in sequences2:
                            distance, _ = fastdtw(seq1, seq2)
                            min_distance = min(min_distance, distance)
                    distance_matrix[i, j] = min_distance
                    distance_matrix[j, i] = min_distance

    # Cluster using HDBSCAN with adjusted parameters
    clusterer = hdbscan.HDBSCAN(metric='precomputed', min_cluster_size=2, min_samples=1)
    cluster_labels = clusterer.fit_predict(distance_matrix)
    
    # Calculate the number of clusters
    n_clusters = len(np.unique(cluster_labels[cluster_labels != -1]))

    # If no clusters are detected, adjust the clustering parameters
    if n_clusters < 2:
        print("No clusters detected. Adjusting clustering parameters.")
        clusterer = hdbscan.HDBSCAN(metric='precomputed', min_cluster_size=2, min_samples=1)
        cluster_labels = clusterer.fit_predict(distance_matrix)

# Create a dictionary to map uniq_id to cluster labels
cluster_mapping = dict(zip(uniq_ids, cluster_labels))

# Add 'trajectory_id' to the original DataFrame based on cluster labels
df['trajectory_id'] = df['uniq_id'].map(cluster_mapping)

# Visualize the clusters (optional)
cluster_counts = df.groupby('trajectory_id')['uniq_id'].count()
plt.bar(cluster_counts.index, cluster_counts.values)
plt.xlabel('Cluster (Trajectory ID)')
plt.ylabel('Number of Trajectories')
plt.title('Cluster Sizes')
plt.show()

# Print DataFrame with trajectory IDs
print("DataFrame with Trajectory IDs:")
print(df[['uniq_id', 'frame', 'label', 'trajectory_id']])
```
