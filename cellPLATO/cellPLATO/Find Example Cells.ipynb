### Find cells with defined characteristics

```python
# Get cells that reflect specific changes of interest. 
factor = 'aspect'
mean_std_df = cp.get_cell_mean_variance(lab_dr_df, factor, sortby='std')

# factor = 'area'
# mean_std_df = get_cell_mean_variance(dr_df, factor, sortby='std')

n_samples = 5
tpt_cutoff = 20

for i in range(n_samples):

    this_cell_df = cp.get_specific_cell(mean_std_df, lab_dr_df,i)
    
    # get this cells id
    this_cell_id = cp.get_cell_id(this_cell_df)  
    
    cp.plot_cell_trajectories(this_cell_df, lab_dr_df, dr_method='umap',contour_scale=1/5)   # Recall contour_scale doing nothing 
    
    
    if((len(this_cell_df) > tpt_cutoff) & cp.ANIMATE_TRAJECTORIES):
        
        # Animate this cells trajectory.
        cp.animate_t_window(this_cell_df, lab_dr_df, dr_method='umap', cid=factor+'_'+this_cell_id)
```
